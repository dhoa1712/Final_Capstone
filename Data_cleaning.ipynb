# Create join table from list of unique countries and list of unique dates

```python
country_list =list (df.country.unique())
country_list
```

```python
data_country = {'country':country_list}
country_df = pd.DataFrame(data_country)
country_df
```

```python
date_list = list(df.date.unique())
date_list
```

```python
data_date ={'date': date_list}
date_df = pd.DataFrame(data_date)
date_df
```

```python
#create key and join on key:
country_df['key'] = 0
date_df['key'] = 0

country_date_df = country_df.merge(date_df, on='key', how='outer')
country_date_df = country_date_df[['country','date']]
```

# Join the newly created table with origional table

```python
new_df = pd.merge(country_date_df, df,  how='left', left_on=['country','date'], right_on = ['country','date'])
```

```python
new_df
```

```python
#sort to make sure country is ascending and date is ascending
new_df = new_df.sort_values(["country", "date"], ascending = (True, True))
new_df = new_df.reset_index(drop=True)
new_df
```

# Clean each column!!

### 1. Clean total_vaccination column


```python
total_vaccinations_cleaned = [0]
for x in range(1,len(new_df)):
    a = new_df.total_vaccinations[x]
    b = new_df.daily_vaccinations[x]
    if new_df.country[x]==new_df.country[x-1]:
        if math.isnan(a) and math.isnan(b):
            tvc = total_vaccinations_cleaned[x-1]
        elif math.isnan(a):
            tvc = total_vaccinations_cleaned[x-1] + b
        else:
            tvc = new_df.total_vaccinations[x]
    else:
        if math.isnan(a):
            tvc = 0
        else:
            tvc = new_df.total_vaccinations
    total_vaccinations_cleaned.append(tvc)
            
```

```python
new_df['total_vaccinations_cleaned']=total_vaccinations_cleaned
```

### 2. Clean people_vaccinated column
